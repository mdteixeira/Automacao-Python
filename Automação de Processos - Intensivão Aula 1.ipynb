```python
import pyautogui as pag # importa e cria um alias para a biblioteca
import pandas as pd
import pyperclip
import time 

pag.PAUSE = 0.5 # define um tempo de pause entre as ações do pyautogui
```

```python
# passo 1 - Entrar no sistema da empresa
pag.hotkey("ctrl", "t") #abre uma nova guia
pag.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pag.press("enter") 
```

```python
# passo 2 - acessar sistema com usuário e senha
time.sleep(1.25) # espera por 1 segundo
pag.press("tab")
pag.write("Username")
pag.press("tab")
pag.write("Senha de exemplo")

pag.press('tab')
pag.press('enter')
```

```python
# passo 3 - baixar os dados

time.sleep(5)
pag.press('j')

pag.press(['tab','tab'])
pag.press('enter')
pag.press(['up','up','up'])
pag.press('enter')
```

```python
# passo 4 - Calcular os indicadores
time.sleep(2) # espera por 2 segundos para o arquivo baixar
tabela = pd.read_csv(r'C:\Users\mathe_bmxf9kl\Downloads\Compras.csv', sep=";") # local do arquivo

display(tabela) # mostra a tabela

# calculos
totalGasto = tabela['ValorFinal'].sum() # soma todos os itens da coluna ValorFinal
quantidade = tabela['Quantidade'].sum() # soma todos os itens da coluna Quantidade
precoMedio = totalGasto / quantidade
```

```python
# passo 5 - Enviar o email

pag.hotkey('ctrl', 't') # abre uma nova guia
pag.write("https://mail.google.com/mail/u/0/#inbox") # acessa o Gmail
pag.press('enter')

time.sleep(5)

pag.press('c') # para funcionar, os atalhos devem estar ativados no Gmail
time.sleep(1)
# preencher as informações do email
# destinatário
pag.write("matheusdiegoteixeira@gmail.com")
pag.press("tab") # escolher o destinatario

pag.press("tab") # passar para o campo assunto
pyperclip.copy("Relatório de Vendas")
pag.hotkey("ctrl", "v")

pag.press("tab") # passando para o corpo do email

texto = f"""
Prezados,
Segue o relatório de compras

Total Gasto: R${totalGasto:,.2f}
Quantidade de Produtos: {quantidade:,}
Preço Médio: R${precoMedio:,.2f}

Qualquer dúvida, é só falar.
Att.,
Matheus Teixeira 
"""

pyperclip.copy(texto)
pag.hotkey("ctrl", "v")


# enviar
pag.hotkey("ctrl", "enter")
```
